```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('RF.csv')
data.head()
```

```python
class_count = data["stu_group"].value_counts()

class_count
```

```python
X = data.drop(columns=["stu_group","Unnamed: 0"])
y = data["stu_group"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.shape
```

```python
# Feature Scalling
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler to the training data and transform both training and testing data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled.shape
```

# **RandomForestClassifier**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Train Decision Tree Classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Make predictions
y_pred = rf.predict(X_test)

# Classification report
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy)
report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)
```

```python
import pandas as pd

# Updated single data point (12 values: consistent with your feature list)
newdata = [16, 1, 5, 3, 85, 80, 75, 70, 85, 78, 388, 77.6]  # Ensure 12 elements

# Group mapping (for readability)
group_mapping = {0: "Arts", 1: "Commerce", 2: "Science"}

def recommend(data):
    # Create DataFrame from the input list
    pre_data = pd.DataFrame({
        'age': [data[0]],
        'location': [data[1]],  # Assume location is already encoded as 1 (e.g., 'Urban')
        'family_size': [data[2]],
        'studytime': [data[3]],
        'attendance': [data[4]],
        'english': [data[5]],
        'math': [data[6]],
        'science': [data[7]],
        'social_science': [data[8]],
        'art_culture': [data[9]],
        'total_score': [data[10]],
        'average_score': [data[11]]
    })

    # Predict group
    predicted_group = rf.predict(pre_data)[0]
    predicted_group_label = group_mapping[predicted_group]

    # Predict probabilities
    probabilities = rf.predict_proba(pre_data)[0]
    probability_dict = {
        "Arts": round(probabilities[0] * 100, 2),
        "Commerce": round(probabilities[1] * 100, 2),
        "Science": round(probabilities[2] * 100, 2)
    }

    return predicted_group_label, probability_dict

# Call the function
predicted_label, probabilities = recommend(newdata)

# Display results
print("Predicted Group:", predicted_label)
print("Prediction Probabilities:")
for group, probability in probabilities.items():
    print(f"{group}: {probability}%")
```
